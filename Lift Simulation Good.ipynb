## Lift Simulation

```python
# created a class
class Lift:
    def __init__(self, current_floor):      # Added constructor __init__
        self.current_floor = current_floor

    def move(self, target_floor):  # added a method to move the lift
        if target_floor > self.current_floor:  # added UP movement logic
            print("Lift is going UP ‚¨ÜÔ∏è")
            for floor in range(self.current_floor + 1, target_floor + 1):
                print("Lift reached floor:", floor)

        elif target_floor < self.current_floor:  # added DOWN movement logic
            print("Lift is going DOWN ‚¨áÔ∏è")
            for floor in range(self.current_floor - 1, target_floor - 1, -1):
                print("Lift reached floor:", floor)

        else:  # added Same floor condition
            print("You are already on this floor!")

        self.current_floor = target_floor # floor updation
        print("Lift has arrived at destination üö™")

# Created a lift object
lift = Lift(int(input("Enter current floor (0-4): ")))

# added Destination floor
target = int(input("Enter destination floor (0-4): "))

# Move the lift
lift.move(target)
```
